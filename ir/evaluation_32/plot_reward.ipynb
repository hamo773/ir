```python
import numpy as np
import matplotlib.pyplot as plt 
import statistics

path = './output_all.txt'

with open(path, 'r') as f:
    lines = f.readlines()

all_rewards = []
for line in lines[1:-1]:
    reward_list = eval(line.strip())
    all_rewards.append(reward_list)
    
#print(all_rewards[1])
std_dev_list = [statistics.stdev(reward_list) for reward_list in all_rewards[1:]]
path = 'sd_32node.txt'
f = open(path, 'w')
f.writelines(str(std_dev_list))
f.close()
#sum_list = [sum(reward_list) for reward_list in all_rewards]
#print(sum_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt 


def get_smooth(l, s_v):
    s_l = []
    for i in range(len(l)):
        s_l.append(l[i] * (1 - s_v) + s_l[i-1] * s_v if i > 0 else l[i])
    return s_l

plt.rcParams.update({'font.size': 14})

if __name__ == '__main__':
    f = open('./output.txt', 'r')
    reward = []
    plt_r = []
    x = []
    for line in f.readlines():
        reward.append(line)
    f.close
    str1 = str(reward).split(',')
    for item in str1[1:-1]:
        plt_r.append((float(item)/992))

    plt.figure(figsize=(10,10))
    plt.xlabel('Step')
    plt.ylabel('Average reward')
    #plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.2, color='blue')  
    #plt_r = get_smooth(plt_r,0.5)
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.2, color='C0')
    plt_r = get_smooth(plt_r,0.95)
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.5, color='blue')  
    plt.savefig('reward.png', dpi=200, bbox_inches='tight')
    plt.show()
```
